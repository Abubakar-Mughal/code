```python
# There are 2 way to fill data 
# one also know as forward/backward fill use in numreic data in dataset and second is mode method that use in       
# f mean forward and fill mean fil its works colomn wise
dataset.fillna(method="ffill")
```

```python
# f mean forward and fill mean fil its works row wise (axis=1)
dataset.fillna(method="ffill",axis=1)
```

```python
# B mean backward and fill mean fil its works row wise (axis=1)
dataset.fillna(method="bfill",axis=1)
```

```python
# Data handle using mode method
mode_data=dataset['Unit Cost'].mode()[0]
print(mode_data)
dataset["Unit Cost"].fillna(mode_data,inplace=True)
dataset.head(5)
```

```python
# Find datatype object
dataset.select_dtypes("object").isnull().sum()
```

```python
# Fill node value value in all object(/categorical data) colomn
for i in dataset.select_dtypes("object").columns:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
dataset.head(5)
```
