```python
import pandas as pd 
dataset=pd.DataFrame({
    "size":["s","m","l","xl","s","m","l","s","s","l","xl"]
})
dataset.head(3)
```

```python

from sklearn.preprocessing import OrdinalEncoder
obj=OrdinalEncoder(categories=[["s","m","l","xl"]])
oe=obj.fit_transform(dataset[["size"]])
dataset["convert"]=oe
dataset
```
