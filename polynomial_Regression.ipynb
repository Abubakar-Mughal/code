```python
import pandas as pd
dataset=pd.read_csv(r"polynomial regression.csv")
dataset.head(5)
```

```python
# Seperate the data for graph and multiple use
x=dataset[["Height (X) in cm"]]
y=dataset["Jump Distance (y) in cm"]
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```python
# Polynomial regression is appy on data that no follow the straight line it mean it follow curved
# degree is power of x
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
poly_feature=PolynomialFeatures(degree=2)
x_poly=poly_feature.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_poly,y)
model.score(x_test,y_test)
```

```python
# Remember when we enter new value then first it convert into polynomial feature using polynomial function
x_new_poly=poly_feature.fit_transform([[20]])
y_predict=model.predict(x_poly)
y_predict
```
